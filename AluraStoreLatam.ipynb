### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
Ingreso_tienda1 = sum(tienda1["Precio"])
Ingreso_tienda2 = sum(tienda2["Precio"])
Ingreso_tienda3 = sum(tienda3["Precio"])
Ingreso_tienda4 = sum(tienda4["Precio"])
```

```python
print(f"El ingreso total de la tienda 1 fue de ${Ingreso_tienda1}")
print(f"El ingreso total de la tienda 2 fue de ${Ingreso_tienda2}")
print(f"El ingreso total de la tienda 3 fue de ${Ingreso_tienda3}")
print(f"El ingreso total de la tienda 4 fue de ${Ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt

# Función para contar y graficar productos por categoría
def contar_y_graficar(df, nombre_tienda):
    categoria_col = 'Categoría del Producto'

    # Contar categorías
    conteo = df[categoria_col].value_counts()

    # Mostrar resultados por consola
    print(f"\nCategorías más populares en {nombre_tienda}:\n")
    print(conteo)

    # Gráfica
    plt.figure(figsize=(10, 6))
    conteo.plot(kind='bar', color='coral')
    plt.title(f"Categorías más populares en {nombre_tienda}")
    plt.xlabel("Categoría")
    plt.ylabel("Cantidad de productos vendidos")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Aplicar a cada tienda
contar_y_graficar(tienda1, "Tienda 1")
contar_y_graficar(tienda2, "Tienda 2")
contar_y_graficar(tienda3, "Tienda 3")
contar_y_graficar(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


Calculando la media de evaluación de cada tienda con base a la calificación de los clientes:

```python
# Función para calcular la media de satisfacción
def calcular_media_satisfaccion(df, nombre_tienda):

    media = df['Calificación'].mean()
    print(f"Media de satisfacción en {nombre_tienda}: {media:.2f}")
    return media

# Calcular la media para cada tienda
media1 = calcular_media_satisfaccion(tienda1, "Tienda 1")
media2 = calcular_media_satisfaccion(tienda2, "Tienda 2")
media3 = calcular_media_satisfaccion(tienda3, "Tienda 3")
media4 = calcular_media_satisfaccion(tienda4, "Tienda 4")
```

Buscamos la categoría mejor evaluada para cada una de las tiendas, esto nos ayudará a saber en que categoría es mejor evalauda cada una de las tiendas:

```python
# Función para encontrar la categoría mejor evaluada
def mejor_categoria(df, nombre_tienda):

    # Agrupar por categoría y calcular media
    promedio_categoria = df.groupby('Categoría del Producto')['Calificación'].mean()

    # Obtener la categoría con mayor promedio
    mejor = promedio_categoria.idxmax()
    puntuacion = promedio_categoria.max()

    print(f"Mejor categoría evaluada en {nombre_tienda}: '{mejor}' con una calificación promedio de {puntuacion:.2f}")
    return mejor, puntuacion

# Aplicar a cada tienda
mejor_categoria(tienda1, "Tienda 1")
mejor_categoria(tienda2, "Tienda 2")
mejor_categoria(tienda3, "Tienda 3")
mejor_categoria(tienda4, "Tienda 4")
```

# 4. Productos más y menos vendidos

```python
# Función para identificar y graficar productos más y menos vendidos
def analizar_ventas_productos(df, nombre_tienda, top_n=5):
    conteo_productos = df['Producto'].value_counts()

    # Identificar más y menos vendidos
    mas_vendidos = conteo_productos.head(top_n)
    menos_vendidos = conteo_productos.tail(top_n)

    print(f"\nTop {top_n} productos más vendidos en {nombre_tienda}:\n{mas_vendidos}")
    print(f"\nTop {top_n} productos menos vendidos en {nombre_tienda}:\n{menos_vendidos}")

    # Gráfico de productos más vendidos
    plt.figure(figsize=(10, 4))
    mas_vendidos.plot(kind='bar', color='seagreen')
    plt.title(f"Top {top_n} productos más vendidos - {nombre_tienda}")
    plt.ylabel("Cantidad de ventas")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

    # Gráfico de productos menos vendidos
    plt.figure(figsize=(10, 4))
    menos_vendidos.plot(kind='bar', color='indianred')
    plt.title(f"Top {top_n} productos menos vendidos - {nombre_tienda}")
    plt.ylabel("Cantidad de ventas")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

# Aplicar a cada tienda
analizar_ventas_productos(tienda1, "Tienda 1")
analizar_ventas_productos(tienda2, "Tienda 2")
analizar_ventas_productos(tienda3, "Tienda 3")
analizar_ventas_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Función para calcular el costo de envío promedio
def calcular_envio_promedio(df, nombre_tienda):
    promedio = df['Costo de envío'].mean()
    print(f"Costo de envío promedio en {nombre_tienda}: ${promedio:.2f}")
    return promedio

# Calcular para cada tienda
envio1 = calcular_envio_promedio(tienda1, "Tienda 1")
envio2 = calcular_envio_promedio(tienda2, "Tienda 2")
envio3 = calcular_envio_promedio(tienda3, "Tienda 3")
envio4 = calcular_envio_promedio(tienda4, "Tienda 4")
```

#6. Ganancias por tienda

```python
# Función para calcular la ganancia
def calcular_ganancia(df):
    return df['Precio'].sum() - df['Costo de envío'].sum()

# Calcular ganancias
ganancia1 = calcular_ganancia(tienda1)
ganancia2 = calcular_ganancia(tienda2)
ganancia3 = calcular_ganancia(tienda3)
ganancia4 = calcular_ganancia(tienda4)

# Mostrar resultados
print(f"La ganancia de la Tienda 1 es de ${ganancia1:,.2f}")
print(f"La ganancia de la Tienda 2 es de ${ganancia2:,.2f}")
print(f"La ganancia de la Tienda 3 es de ${ganancia3:,.2f}")
print(f"La ganancia de la Tienda 4 es de ${ganancia4:,.2f}")

# Pie chart
ganancias = [ganancia1, ganancia2, ganancia3, ganancia4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 6))
plt.pie(ganancias, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title("Distribución porcentual de las ganancias por tienda")
plt.axis('equal')  # Asegura que el pie sea circular
plt.show()
```
